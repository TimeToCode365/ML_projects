If we directly did 'backup = df' this does not create an independent copy.

#### Now we shall replace the numbers present in the Name column with NaN.

```python
import numpy as np
```

Here we imported numpy because if we used pd.NA in the below command it would replace it with `<NA>` ,and not NaN, as the Name column is of string type.But there already exists some NaN values and if `<NA>` values get added,it creates data inconsistency!

It would be better if they are consistent with NaN,as this type already exists in the dataset.

```python
df['Name'].apply(type).value_counts()
```

The above commad was run to see the different datatypes present in this column.

```python
df['Name'] = df['Name'].where(~df['Name'].astype(str).str.isdigit(), np.nan)
```

where() keeps values where a condition is True, and replaces others with NaN (or another value you specify).

The astype(str) first converts the input to string datatype then works on it.

#### The only column left to fix is the email one.

Define a regex(regular expression) pattern for valid emails.(It checks if the email starts with valid username characters, has an '@', followed by a valid domain with a dot, and ends with a domain extension of at least two letters.)

r => Raw string to avoid escape issues

{2,} => Extension with 2+ letters

$ => End of string

```python
valid_email_pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
```

Apply the pattern and replace invalid ones with NaN.

```python
df['Email'] = df['Email'].where(df['Email'].str.match(valid_email_pattern, na=False), np.nan)
```

`str.match` checks if the entire string matches the regex pattern from start to end. It returns True if it matches fully, else False.

na=False means: if the value is missing (NaN), treat it as not matching (i.e., return False) instead of NaN.
